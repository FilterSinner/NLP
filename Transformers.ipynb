```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "Life is beatifully unfair",
    max_length=50,
    num_return_sequences=2,
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> in the world of Advaced AI.", top_k=2)
```

```python
mask = pipeline("fill-mask", model = "bert-base-cased")
mask("Hello there [MASK], I've been a teacher since 10 years now!",top_k = 2)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("I am Ojal Binoj studying in VIT University, doing my computer science degree")
```

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
    """
    Life is a complex tapestry of experiences, a journey filled with both joy and adversity. It's a continuous evolution, a dynamic dance of moments that shape our identity. From the exhilarating highs of love, success, and laughter to the challenging lows of loss, failure, and tears, life offers a mosaic of emotions and lessons. It's a canvas on which we paint our dreams, aspirations, and connections with others. In its fleeting nature, life compels us to savor each instant, to learn, to grow, and to embrace change. Through these myriad encounters, we uncover our purpose, seek meaning, and ultimately discover the profound beauty that exists within our existence.
"""
)
```

```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")
translator("I find Hugging Face interface to be cool")
```
